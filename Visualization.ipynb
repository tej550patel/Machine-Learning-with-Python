```python
plt.figure(figsize=(20,10))
plt.scatter(data=df,x='Attack',y='Defense',s=200,c='green',marker='D')
#change axes ranges
plt.xlim(0,150)
plt.ylim(0,200)
#add title
plt.title('Relationship between Attack and Defense',size=20)
#add x and y labels
plt.xlabel('Attacks',size=15)
plt.ylabel('Defense',size=15)
plt.grid(True)
#show plot
plt.savefig("test.jpg")
plt.show()
```

```python
sns.pairplot(df,x_vars=["Attack","Defense","Speed"],y_vars="Total",
            kinf)
```

```python
# Line Plot
x = [10,30,20,70]
y = [10,45,10,70]
plt.plot(x,y,c="purple")

plt.xlim(0,100)
plt.ylim(0,100)
plt.title('A simple line plot')
plt.grid(True)
plt.savefig("test.jpg")
for i_x, i_y in zip(x,y):
    plt.text(i_x,i_y,(i_x,i_y))
    
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(15,7))
plt.subplot(121)
plt.hist(x='Attack',data=df,bins=10,color='darkblue')
plt.title('Attacks')
plt.subplot(122)
plt.hist(x='Defense',data=df,bins=20,color='darkblue')
plt.title('Defense')
plt.xticks(np.arange(0,201,10))
plt.show()
```

```python
#distribution plot
sns.distplot(df.Attack)
```

```python
sns.countplot(x='Type 2',data=df,palette="Set1")

#Rotate x-labels
plt.xticks(rotation=45)
```

```python
#Count Plot (a.k.a Frequency bar Plot)
sns.countplot(x='Type 2',data=df,palette="Set2",
             order=df['Type 2'].value_counts().sort_index().index)
#rotate x-labels
plt.xticks(rotation=45)
```
